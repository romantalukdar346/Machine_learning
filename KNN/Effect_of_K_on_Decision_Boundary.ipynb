```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
df= pd.read_csv(r'C:\Users\KIIT\OneDrive\Desktop\code 1\Machine_Learning_csv_file\Social_Network_Ads.csv')
```

```python
df
```

```python
x=df.iloc[:,2:4].values
```

```python
y=df.iloc[:,-1].values
```

```python
x.shape,y.shape
```

```python
scaler=StandardScaler()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
training_error=[]
testing_error=[]
for i in range(1,26):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    x=confusion_matrix(y_train,knn.predict(x_train))
    y=confusion_matrix(y_test,knn.predict(x_test))
    training_error.append((x[0][1]+x[1][0])/x.sum())
    testing_error.append((y[0][1]+y[1][0])/y.sum())
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))
plt.title("Effect of K")
plt.plot(range(1,26),training_error,label="Training_Error")
plt.plot(range(1,26),testing_error,label="Test_Error")
plt.xlabel("K values")
plt.ylabel("Error")
plt.legend()
```

```python
def Keffect(data,k):
    df= pd.read_csv(r'C:\Users\KIIT\OneDrive\Desktop\code 1\Machine_Learning_csv_file\{}'.format(data))
    x=df.iloc[:,:2].values
    y=df.iloc[:,-1].values
    
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x,y)
    
    a=np.arange(start=x[:,0].min()-1,stop=x[:,0].max()+1,step=0.01)
    b=np.arange(start=x[:,1].min()-1,stop=x[:,1].max()+1,step=0.01)
    
    xx,yy=np.meshgrid(a,b)
    
    input_arr=np.array([xx.ravel(),yy.ravel()]).T
    label=knn.predict(input_arr)
    
    plt.contourf(xx,yy,label.reshape(xx.shape),alpha=0.7)
    sns.scatterplot(x[:,0],x[:,1],c=y)
```
