# Creating dataset and plotting the data

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
        n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

# training dataset using LogisticRegression library

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(penalty='none',solver='sag')
lor.fit(X,y)
```

```python
print(lor.coef_)
print(lor.intercept_)
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
print(f"{m1}   {b1}")
```

```python

import random
```

# Sigmoid function 

```python

def sigmoid(z):
    return 1/(1+np.exp(-z))
```

# LogisticRegression algo where Gradient Descent  is used

```python
def logicreg(x_train,y_train):
    x_train=np.insert(x_train,0,1,axis=1)
    weight=np.ones(x_train.shape[1])
    lr=0.1
    for j in range(2500):
        
        y_hat=sigmoid(np.dot(x_train,weight))
        weight=weight+(lr*np.dot((y_train-y_hat),x_train))/x_train.shape[0]
    return weight
```

# Object declaration

```python
log_reg=logicreg(X,y)
print(log_reg)
print("coef_      intercept_")
```

```python
inter_2=log_reg[0]
coef_2=log_reg[1:]
```

```python
print(f"{inter_2}   {coef_2}")
```

```python
m3 = -(coef_2[0]/coef_2[1])
b3 = -(inter_2/coef_2[1])
```

```python
print(f"{m3}   {b3}")
```

# Logistic Regression line for both Algorithm

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m1*x_input1 + b1  #  logistic library line

y_input3 = m3*x_input1 + b3  #  logistic sigmoid function line


```

```python
# using logistic library        (black)
# using logistic sigmoid function (yellow)

plt.figure(figsize=(10,6))
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.plot(x_input1,y_input3,color='yellow',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
